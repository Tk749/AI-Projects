```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris') # dataset
```

```python
# Scatter

setosa = iris[iris.species == 'setosa']
versicolor = iris[iris.species == 'versicolor']
virginica = iris[iris.species == 'virginica']

# Create a figure with two subplots, arranged side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot the first subplot (Petal Length vs Width)
axes[0].scatter(x='petal_length', y='petal_width', data=setosa, color='blue', label='Setosa')
axes[0].scatter(x='petal_length', y='petal_width', data=versicolor, color='orange', label='Versicolor')
axes[0].scatter(x='petal_length', y='petal_width', data=virginica, color='green', label='Virginica')
axes[0].set_xlabel("Petal Length")
axes[0].set_ylabel("Petal Width")
axes[0].set_title("Petal Length vs Width")
axes[0].legend()

# Plot the second subplot (Sepal Length vs Width)
axes[1].scatter(x='sepal_length', y='sepal_width', data=setosa, color='blue', label='Setosa')
axes[1].scatter(x='sepal_length', y='sepal_width', data=versicolor, color='orange', label='Versicolor')
axes[1].scatter(x='sepal_length', y='sepal_width', data=virginica, color='green', label='Virginica')
axes[1].set_xlabel("Sepal Length")
axes[1].set_ylabel("Sepal Width")
axes[1].set_title("Sepal Length vs Width")
axes[1].legend()

# Adjust the layout to avoid overlapping labels and titles
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Boxplot

# Set up the subplot layout
fig, axes = plt.subplots(2, 2, figsize=(6, 5))

# Box plot for 'sepal_length' by 'species'
sns.boxplot(data=iris, x='species', y='sepal_length', ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length by Species')

# Box plot for 'sepal_width' by 'species'
sns.boxplot(data=iris, x='species', y='sepal_width', ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width by Species')

# Box plot for 'petal_length' by 'species'
sns.boxplot(data=iris, x='species', y='petal_length', ax=axes[1, 0])
axes[1, 0].set_title('Petal Length by Species')

# Box plot for 'petal_width' by 'species'
sns.boxplot(data=iris, x='species', y='petal_width', ax=axes[1, 1])
axes[1, 1].set_title('Petal Width by Species')

# Adjust the layout to prevent overlapping titles
plt.tight_layout()

# Display the plots
plt.show()
```

```python
# Violinplot

# Set up the subplot layout
fig, axes = plt.subplots(2, 2, figsize=(6, 5))

# Box plot for 'sepal_length' by 'species'
sns.violinplot(data=iris, x='species', y='sepal_length', ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length by Species')

# Box plot for 'sepal_width' by 'species'
sns.violinplot(data=iris, x='species', y='sepal_width', ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width by Species')

# Box plot for 'petal_length' by 'species'
sns.violinplot(data=iris, x='species', y='petal_length', ax=axes[1, 0])
axes[1, 0].set_title('Petal Length by Species')

# Box plot for 'petal_width' by 'species'
sns.violinplot(data=iris, x='species', y='petal_width', ax=axes[1, 1])
axes[1, 1].set_title('Petal Width by Species')

# Adjust the layout to prevent overlapping titles
plt.tight_layout()

# Display the plots
plt.show()
```

# Data

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Separate input features and target labels
X = iris.drop('species', axis=1)  # Input features (remove the 'species' column)
y = iris['species']  # Target labels

# Encode target labels into numerical values
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6,random_state=42) # %40 Train - %60 Test 
```

# Train & Test

```python
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
# Classification Models
models = {
    'Decision Tree'          : tree.DecisionTreeClassifier(),
    'Logistic Regression'    : LogisticRegression(max_iter=1000, random_state=42),
    'Naive Bayes'            : GaussianNB(),
    'K-Nearest Neighbors'    : KNeighborsClassifier(n_neighbors=3),
    'Support Vector Machine' : SVC(kernel='linear', C=1.0, random_state=42)
}

results = []
```

```python
# Model training and calculate of accuracy rate
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results.append({'Model': model_name, 'Accuracy': accuracy})
```

```python
# Results
results_df = pd.DataFrame(results)
results_df 
```
