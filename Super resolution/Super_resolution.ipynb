```python
import cv2
from cv2 import dnn_superres
import matplotlib.pyplot as plt
```

```python
# Create an SR object
sr = dnn_superres.DnnSuperResImpl_create()

# Read image
image = cv2.imread('./images.png')
```

```python
path = "EDSR_x4.pb"
sr.readModel(path)

# Set the desired model and scale to get correct pre- and post-processing
sr.setModel("edsr", 4)
```

```python
result = sr.upsample(image)

# Save the upscaled image
cv2.imwrite("./upscaled.png", result)
```

```python
image1 = cv2.imread('images.png')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)  # Fix OpenCV's default color scheme

image2 = cv2.imread('upscaled.png')
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)  # Fix OpenCV's default color scheme

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.imshow(image1)
ax1.set_title('Original Image')

ax2.imshow(image2)
ax2.set_title('Upscaled Image')

ax1.axis('off')
ax2.axis('off')

plt.show()
```
